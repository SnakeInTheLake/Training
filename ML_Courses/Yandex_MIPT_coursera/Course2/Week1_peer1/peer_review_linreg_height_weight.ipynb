Это можно делать с помощью метода $scatter\_matrix$ Pandas Data Frame или *pairplot* библиотеки Seaborn. 

Чтобы проиллюстрировать этот метод, интересней добавить третий признак. Создадим признак *Индекс массы тела* ([BMI](https://en.wikipedia.org/wiki/Body_mass_index)). Для этого воспользуемся удобной связкой метода *apply* Pandas DataFrame и lambda-функций Python.

```python
def make_bmi(height_inch, weight_pound):
    METER_TO_INCH, KILO_TO_POUND = 39.37, 2.20462
    return (weight_pound / KILO_TO_POUND) / \
           (height_inch / METER_TO_INCH) ** 2
```

```python
data['BMI'] = data.apply(lambda row: make_bmi(row['Height'], 
                                              row['Weight']), axis=1)
```

## Задание 2. Минимизация квадратичной ошибки

В простейшей постановке задача прогноза значения вещественного признака по прочим признакам (задача восстановления регрессии) решается минимизацией квадратичной функции ошибки. 

**1.** Напишите функцию, которая по двум параметрам $w_0$ и $w_1$ вычисляет квадратичную ошибку приближения зависимости роста $y$ от веса $x$ прямой линией $y = w_0 + w_1 * x$:
$$error(w_0, w_1) = \sum_{i=1}^n {(y_i - (w_0 + w_1 * x_i))}^2 $$
Здесь $n$ – число наблюдений в наборе данных, $y_i$ и $x_i$ – рост и вес $i$-ого человека в наборе данных. 

```python
def error(w_0, w_1):
    '''
    Возвращает квадратичную ошибку, соответствующую определённым w_0 и w_1
    '''
    x = data['Weight']
    y = data['Height']
    err = y - (w_0 + w_1 * x)
    return (err**2).sum()
```

Итак, мы решаем задачу: как через облако точек, соответствующих наблюдениям в нашем наборе данных, в пространстве признаков "Рост" и "Вес" провести прямую линию так, чтобы минимизировать функционал из п. 1. Для начала давайте отобразим хоть какие-то прямые и убедимся, что они плохо передают зависимость роста от веса.

**2.** Проведите на графике из п. 4 Задания 1 две прямые, соответствующие значениям параметров ($w_0, w_1) = (60, 0.05)$ и ($w_0, w_1) = (50, 0.16)$. Используйте метод *plot* из *matplotlib.pyplot*, а также метод *linspace* библиотеки NumPy. Подпишите оси и график.

```python
def line(w_0, w_1, x):
    '''
    Возвращает прямую, соответствующую определённым w_0 и w_1
    '''
    return w_0 + w_1 * x
```

```python
xx = np.linspace(70, 170, 100) # Разметка оси x
data.plot(x='Weight', y='Height', kind='scatter', title='Зависимость роста от веса') # График из п.5
plt.xlabel('Вес')
plt.ylabel('Рост')
plt.plot(xx, line(60, 0.05, xx), 'r', label='w_0 = 60; w_1 = 0.05')  # Прямая, соответствующая w_0 = 60 и w_1 = 0.05
plt.plot(xx, line(50, 0.16, xx), 'c', label='w_0 = 50; w_1 = 0.16')  # Прямая, соответствующая w_0 = 50 и w_1 = 0.16
plt.legend();
```

Минимизация квадратичной функции ошибки - относительная простая задача, поскольку функция выпуклая. Для такой задачи существует много методов оптимизации. Посмотрим, как функция ошибки зависит от одного параметра (наклон прямой), если второй параметр (свободный член) зафиксировать.

**3.** Постройте график зависимости функции ошибки, посчитанной в п. 1, от параметра $w_1$ при $w_0$ = 50. Подпишите оси и график.

```python
xx = np.linspace(-1, 1, 100)
yy = [error(50, w_1) for w_1 in xx]
plt.plot(xx, yy)
plt.title('Зависимость ошибки от w_1 при w_0 = 50')
plt.xlabel('w_1')
plt.ylabel('Ошибка');
```

Теперь методом оптимизации найдем "оптимальный" наклон прямой, приближающей зависимость роста от веса, при фиксированном коэффициенте $w_0 = 50$.

**4.** С помощью метода *minimize_scalar* из *scipy.optimize* найдите минимум функции, определенной в п. 1, для значений параметра  $w_1$ в диапазоне [-5,5]. Проведите на графике из п. 4 Задания 1 прямую, соответствующую значениям параметров ($w_0$, $w_1$) = (50, $w_1\_opt$), где $w_1\_opt$ – найденное в п. 3 оптимальное значение параметра $w_1$. 

```python
error_part = partial(error, 50) # Фиксирует в функции ошибки w_0 = 50
w_1_opt = minimize_scalar(error_part, bounds=(-5, 5), method='bounded')
print(f'Оптимальный w_1 = {round(w_1_opt.x, 3)}')
```

```python
xx = np.linspace(70, 170, 100) # Разметка оси x
data.plot(x='Weight', y='Height', kind='scatter', title='Зависимость роста от веса') # График из п.5
plt.xlabel('Вес')
plt.ylabel('Рост')
plt.plot(xx, line(50, w_1_opt.x, xx), 'r', label='w_0 = 50; w_1 = %.3f' % w_1_opt.x)  # Прямая, соответствующая w_0 = 50 и оптимальному w_1
plt.legend();
```

**5.** Постройте 3D-график зависимости функции ошибки, посчитанной в п.1 от параметров $w_0$ и $w_1$. Подпишите ось $x$ меткой «Intercept», ось $y$ – меткой «Slope», a ось $z$ – меткой «Error».

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.gca(projection='3d') # get current axis

# Создаем массивы NumPy с координатами точек по осям X и У. 
# Используем метод meshgrid, при котором по векторам координат 
# создается матрица координат. Задаем нужную функцию Z(x, y).
X = np.arange(-100, 100, 1)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)

# Заполняем матрицу Z значениями функции ошибки
Z = np.zeros(X.shape)
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        Z[i, j] = error(X[i, j], Y[i, j])

# Наконец, используем метод *plot_surface* объекта 
# типа Axes3DSubplot. Также подписываем оси.
surf = ax.plot_surface(X, Y, Z)
ax.set_xlabel('Intercept')
ax.set_ylabel('Slope')
ax.set_zlabel('Error')
plt.show()
```

**6.** С помощью метода *minimize* из scipy.optimize найдите минимум функции, определенной в п. 1, для значений параметра $w_0$ в диапазоне [-100,100] и $w_1$ - в диапазоне [-5, 5]. Начальная точка – ($w_0$, $w_1$) = (0, 0). Используйте метод оптимизации L-BFGS-B (аргумент method метода minimize). Проведите на графике из п. 4 Задания 1 прямую, соответствующую найденным оптимальным значениям параметров $w_0$ и $w_1$. Подпишите оси и график.

Функция minimize требует, чтобы аргументы функции, которую нужно оптимизировать, были в виде одномерного массива формы (n, ), где n - количество аргументов. Поэтому ниже перепишем функцию так, чтобы она принимала на вход массив, а остальное оставим без изменений.

```python
def error_arr(ws):
    x = data['Weight']
    y = data['Height']
    err = y - (ws[0] + ws[1] * x)
    return (err**2).sum()
```

```python
x0 = [0, 0]
bounds = [(-100, 100), (-5, 5)]
opt = minimize(error_arr, x0=x0, method='L-BFGS-B', bounds=bounds)
opt_w0, opt_w1 = opt.x
print(f'Оптимальный w_0 = {round(opt_w0, 3)}\nОптимальный w_1 = {round(opt_w1, 3)}')
```

```python
xx = np.linspace(70, 170, 100) # Разметка оси x
data.plot(x='Weight', y='Height', kind='scatter', title='Зависимость роста от веса') # График из п.5
plt.xlabel('Вес')
plt.ylabel('Рост')
plt.plot(xx, line(opt_w0, opt_w1, xx), 'r', label='w_0 = %.3f; w_1 = %.3f' % (opt_w0, opt_w1))  # Прямая, соответствующая оптимальным w_0 и w_1
plt.legend();
```
